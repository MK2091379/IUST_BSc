```python
import mahotas
import cv2
import numpy as np
# loading image1
img1 = cv2.imread('img1.png')
img2 = cv2.imread('img2.png')
img3 = cv2.imread('img3.png')
img4 = cv2.imread('img4.png')
img5 = cv2.imread('img5.png')
# filtering image
img1 = img1[:, :, 0]
img2 = img2[:, :, 0]
img3 = img3[:, :, 0]
img4 = img4[:, :, 0]
img5 = img5[:, :, 0]
# computing roundness in image
value1 = mahotas.features.roundness(img1)
value2 = mahotas.features.roundness(img2)
value3 = mahotas.features.roundness(img3)
value4 = mahotas.features.roundness(img4)
value5 = mahotas.features.roundness(img5)
# printing value
print("Roundness of img1 : " + str(value1))
print("Roundness of img2 : " + str(value2))
print("Roundness of img3 : " + str(value3))
print("Roundness of img4 : " + str(value4))
print("Roundness of img5 : " + str(value5))
print("_____________________________________________________")
# computing eccentricity in image
value1 = mahotas.features.eccentricity(img1)
value2 = mahotas.features.eccentricity(img2)
value3 = mahotas.features.eccentricity(img3)
value4 = mahotas.features.eccentricity(img4)
value5 = mahotas.features.eccentricity(img5)
# printing value
print("eccentricity of img1 : " + str(value1))
print("eccentricity of img2 : " + str(value2))
print("eccentricity of img3 : " + str(value3))
print("eccentricity of img4 : " + str(value4))
print("eccentricity of img5 : " + str(value5))
print("_____________________________________________________")
ret1,thresh1 = cv2.threshold(img1,127,255,0)
ret2,thresh2 = cv2.threshold(img2,127,255,0)
ret3,thresh3 = cv2.threshold(img3,127,255,0)
ret4,thresh4 = cv2.threshold(img4,127,255,0)
ret5,thresh5 = cv2.threshold(img5,127,255,0)
contours1,hierarchy1 = cv2.findContours(thresh1, 1, 2)
contours2,hierarchy2 = cv2.findContours(thresh2, 1, 2)
contours3,hierarchy3 = cv2.findContours(thresh3, 1, 2)
contours4,hierarchy4 = cv2.findContours(thresh4, 1, 2)
contours5,hierarchy5 = cv2.findContours(thresh5, 1, 2)
cnt1 = contours1[0]
cnt2 = contours2[0]
cnt3 = contours3[0]
cnt4 = contours4[0]
cnt5 = contours5[0]
perimeter1 = cv2.arcLength(cnt1,True)
perimeter2 = cv2.arcLength(cnt2,True)
perimeter3 = cv2.arcLength(cnt3,True)
perimeter4 = cv2.arcLength(cnt4,True)
perimeter5 = cv2.arcLength(cnt5,True)
area1 = cv2.contourArea(cnt1)
area2 = cv2.contourArea(cnt2)
area3 = cv2.contourArea(cnt3)
area4 = cv2.contourArea(cnt4)
area5 = cv2.contourArea(cnt5)
compactness1=(perimeter1**2)/area1
compactness2=(perimeter2**2)/area2
compactness3=(perimeter3**2)/area3
compactness4=(perimeter4**2)/area4
compactness5=(perimeter5**2)/area5
print("compactness of img1 : " + str(compactness1))
print("compactness of img2 : " + str(compactness2))
print("compactness of img3 : " + str(compactness3))
print("compactness of img4 : " + str(compactness4))
print("compactness of img5 : " + str(compactness5))
```
