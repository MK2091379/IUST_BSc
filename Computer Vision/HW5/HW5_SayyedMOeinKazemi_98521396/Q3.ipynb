```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import math
from google.colab import drive
```

```python
img = cv2.imread('Scientist.jpg', 0)
```

```python
# Calculate and show histogram #
  ### Start your code ###
hist = cv2.calcHist([img],[0],None,[256],[0,256])
hist
  ### End ###
```

```python
def global_thresholding(image , k):
  ### Start your code ###
  return cv2.threshold(image,k,255,cv2.THRESH_BINARY)[1]
  ### End ###
```

```python
def otsu_thresholding(image):
  ### Start your code ###
  image = image.astype("uint8")
  return cv2.threshold(image,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)[1]
  ### End ###
```

```python
def adaptive_thresholding(image , C , blocksize):
  ### Start your code ###
  return cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, C, blocksize)
  ### End ###
```

```python
# try a few k's to find best k 
# set your parameters #
k = 127
image_global = global_thresholding(img , k)
image_otsu = otsu_thresholding(img)
image_adaptive = adaptive_thresholding(img , C = 11, blocksize = 2)

# show and compare results here
# dont change this part
plt.gray()
fig = plt.figure(figsize=(20,20))
cols = 4
rows = 1

fig.add_subplot(rows , cols , 1)
plt.imshow(img)
plt.axis('off')
plt.title("Original image")

fig.add_subplot(rows , cols , 2)
plt.imshow(image_global)
plt.axis('off')
plt.title("Global Thresholding image")

fig.add_subplot(rows , cols , 3)
plt.imshow(image_otsu)
plt.axis('off')
plt.title("Otsu Thresholding image")

fig.add_subplot(rows , cols , 4)
plt.imshow(image_adaptive)
plt.axis('off')
plt.title("Adaptive Thresholding image")

plt.show()
```
