```python
#import libraries
import matplotlib.pyplot as plt
import cv2
#function to plot histogram
def histogram(img):
  pixels=[]

  #create list of values 0-255
  for x in range(256):
    pixels.append(x)
  
  #initialize width and height of image
  width,height=img.shape
  counts=[]
  
  #for each intensity value
  for i in pixels:
    #set counter to 0
    temp=0

    #traverse through the pixels
    for x in range(width):
      for y in range(height):

        #if pixel intensity equal to intensity level
        #increment counter
        if (img[x][y]==i):
          temp=temp+1

    #append frequency of intensity level 
    counts.append(temp)
  
  #plot histogram
  plt.bar(pixels,counts)
  plt.show()
```

```python
img = cv2.imread('img1.jpg',cv2.IMREAD_GRAYSCALE)
histogram(img)
```
